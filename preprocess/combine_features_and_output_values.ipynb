```python
import pandas as pd
```

```python
Flourishing_result = pd.read_csv('../processed_output/Flourishing_result.csv')
```

```python
Flourishing_result[:5]
```

```python
Flourishing_result = Flourishing_result.drop(columns = ['Flourishing_score'])
Flourishing_result[:5]
```

```python
features = pd.read_csv('data/all_features_without_outliers.csv')
features = features.iloc[:,1:]
features = features.fillna(0)
features[:5]
```

```python
features.shape
```

```python
def get_line(data):
    min_num = min(data)
    max_num = max(data)
    new_data = []
    for i in data:
        new_one = (max_num - i)/(max_num - min_num)
        new_data.append(new_one)
    return new_data

def normalization(all_data):
    #print(all_data.columns.values)
    for i in range(1,len(all_data.columns.values)):
        #print(all_data.columns.values[i])
        all_data[all_data.columns.values[i]] = get_line(all_data[all_data.columns.values[i]])
    return all_data
```

```python
normalised = normalization(features)
normalised[:5]
```

```python
features_and_flourishing_values = normalised.merge(Flourishing_result, on = 'uid')
features_and_flourishing_values.shape
```

```python
features_and_flourishing_values.to_csv('data/features_and_flourishing_values.csv')
```
