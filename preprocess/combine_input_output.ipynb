```python
import pandas as pd
```

```python
features = pd.read_csv('data/all_features_normalised.csv')
flourishing = pd.read_csv('../processed_output/Flourishing_result_without0_label.csv')
panas = pd.read_csv('../processed_output/panas_result_without0_label.csv')
```

```python
features = features.iloc[:,1:]
features[:5]
```

```python
flourishing = flourishing.iloc[:,1:]
flourishing[:5]
```

```python
panas = panas.iloc[:,1:]
panas[:5]
```

```python
features_and_flourishing = pd.merge(features, flourishing, on='uid', how='inner')
features_and_flourishing.to_csv('data/features_and_flourishing.csv')
features_and_flourishing
```

```python
features_and_panas = pd.merge(features, panas, on='uid', how='inner')
features_and_panas.to_csv('data/features_and_panas.csv')
features_and_panas
```
